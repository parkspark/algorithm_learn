# 왜 효율적인 알고리즘을 사용해야 하는가?

### 1. 빠른 속도를 위해서
동일한 문제를 해결하더라도 시간이 다르다.
무어의 법칙처럼 cpu나 하드웨어가 2년마다 2배씩 성능이 증가한다면 빨리돌아가는 알고리즘을 고민할 시간에 새로운 cpu를 얻는게 더 나을수도 있다.
하지만 하드웨어 발전속도는 점점 느려졌고, 더 빠른 알고리즘을 고민하는 것이 매우 중요해 졌다.
시간 복잡도(time complexity)의 개념이 도입된다.
어떠한 크기의 입력값(n)에 대해서 알고리즘을 수행하는 동안 몇 번의 연산을 실행하는지 나타낸다.

### 2. 적은 공감
알고리즘의 효율성을 평가하는데 얼마만큼의 메모리를 사용하느냐? 도 있다.

1 부터 2^32 까지의 덧셈을 수행하는데 조금 특이한 방식으로 진행한다.

큰 배열을 생성한 다음, 첫 번째 원소에는 그냥 1, 두 번째 원소에는 1 부터 2 까지 더한거, 세 번째 원소에는 1 부터 3 까지 더한거, ... 쭉 해서 
n 번째 원소에는 1 부터 n 까지 더한것을 저장한다.

그런 다음에, 사용자가 임의의 n 에 대해서 1 부터 n 까지 더한 값을 요청한다면, 덧셈을 새로 수행할 필요 없이 그냥 배열의 n 번째 원소를 리턴해주면 됩니다. 
만약에, 앞서 배열을 생성하는 작업을 무시한다면 (아니면 남이 만들어놓은 배열을 그냥 가져다 쓴다고 생각하셔도 됩니다), 
사용자에 요청에 대해서 단 한 번의 연산 (배열의 n 번째 원소를 가져옴)으로 덧셈을 수행할 수 있습니다.

즉, 이 알고리즘의 시간 복잡도는 단 1 이 되겠지요. 그렇다면 이 알고리즘은 효율적인 것일까요? 
아마 아무도 그렇게 생각하지 않을 것입니다. 왜냐하면 크기가 무려 2^32, 즉 4 기가 바이트에 달하는 배열이 필요하기 떄문.

이러한 문제 때문에 공간 복잡도(Space complexity)를 효율성을 판단하는데 중요한 요소이다.

```python
# 위의 공간 복잡도에 예시를 들은 1부터 2^32 까지의 배열로 합을 구하는 코드.
# 절대 실행하지 말 것. 매우 많은 메모리를 잡아먹음. 
# DO NOT RUN THIS CODE
'''
n = 2**32
cumulative_sum = [0] * (n + 1)
for i in range(1, n + 1):
    cumulative_sum[i] = cumulative_sum[i - 1] + i
'''
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import math

# data 생성
n = np.linspace(1, 10, 100)
n_square = n**2
n_log_n = n * np.log(n)
n_2_power_n = 2**n

# 데이터를 저장한 dictionary 생성
data = {'N': n,
        'N^2': n_square,
        'N log N': n_log_n,
        '2^N': n_2_power_n}


# 그리기 위해 Dataframe으로 변환
df = pd.DataFrame(data)

# 데이터 시각화
sns.lineplot(data=df, palette="tab10", linewidth=2.5)
plt.xlabel('N')
plt.ylabel('Y')
plt.title('Big-O Complexity Chart')
plt.show()
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt

# data 생성
n = np.arange(1, 7)
n_factorial = [math.factorial(x) for x in n]

# 데이터 시각화
plt.plot(n, n_factorial, marker='o', linestyle='-')
plt.xlabel('N')
plt.ylabel('N!')
plt.title('Factorial function N!')

# y축 스케일을 자연수로 설정
plt.yscale("linear")

plt.show()
```
